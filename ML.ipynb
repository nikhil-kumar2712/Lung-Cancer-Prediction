```python
from sklearn.linear_model import LogisticRegression
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score
#Fitting simple linear regression to the training test
Model1 = LogisticRegression()
Model1.fit(X_train, Y_train)
#Predicting the test set results
Y_pred_LR = Model1.predict(X_test)
cm = confusion_matrix(Y_test, Y_pred_LR)
accuracy = accuracy_score(Y_test, Y_pred_LR)

print("Confusion Matrix:\n", cm)
print("Accuracy Score:", accuracy)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Oranges', cbar=False)
plt.title(f"Confusion Matrix (Accuracy: {accuracy:.2f})")
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import KFold
import numpy as np

dt = DecisionTreeClassifier(random_state=0)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

cms = []  # List to store confusion matrices for each fold
accuracies = []  # List to store accuracies for each fold

for i, (train_index, test_index) in enumerate(kf.split(X_train)):
    X_train_kf, X_val_kf = X_train.iloc[train_index], X_train.iloc[test_index]
    y_train_kf, y_val_kf = Y_train[train_index], Y_train[test_index]
    dt.fit(X_train_kf, y_train_kf)
    Y_pred_dt = dt.predict(X_val_kf)
    cm = confusion_matrix(y_val_kf, Y_pred_dt)
    accuracy = accuracy_score(y_val_kf, Y_pred_dt)
    cms.append(cm)
    accuracies.append(accuracy)

# Calculate average confusion matrix and accuracy
avg_cm = np.mean(cms, axis=0)
avg_accuracy = np.mean(accuracies)

# Plot the average confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(avg_cm, annot=True, fmt='.2f', cmap='Greens', cbar=False)
plt.title(f"Average Confusion Matrix (Accuracy: {avg_accuracy:.2f})")
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier

k = 3  # Number of neighbors
knn = KNeighborsClassifier(n_neighbors=k)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Loop over each fold
for i, (train_index, test_index) in enumerate(kf.split(X_train)):
    X_train_kf, X_val_kf = X_train.iloc[train_index], X_train.iloc[test_index]
    y_train_kf, y_val_kf = Y_train[train_index], Y_train[test_index]
    knn.fit(X_train_kf, y_train_kf)
    Y_pred_knn = knn.predict(X_val_kf)
    cm = confusion_matrix(y_val_kf, Y_pred_knn)
    accuracy = accuracy_score(y_val_kf, Y_pred_knn)
    cms.append(cm)
    accuracies.append(accuracy)

# Calculate average confusion matrix and accuracy
avg_cm = np.mean(cms, axis=0)
avg_accuracy = np.mean(accuracies)


# Plot the average confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(avg_cm, annot=True, fmt='.2f', cmap='Blues', cbar=False)
plt.title(f"Average Confusion Matrix (Accuracy: {avg_accuracy:.2f})")
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

model = Sequential([
    Input(shape=(X_train.shape[1],)),
    Dense(10, activation='relu'),
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X_train, Y_train, epochs=50, batch_size=5, verbose=1)

Y_pred_prob = model.predict(X_test)
Y_pred_nn = (Y_pred_prob > 0.5).astype(int)

cm_nn = confusion_matrix(Y_test, Y_pred_nn)
print("Confusion Matrix:\n", cm_nn)
accuracy = accuracy_score(Y_test, Y_pred_nn)
print(f"Accuracy: {accuracy:.2f}")

plt.figure(figsize=(8, 6))
sns.heatmap(cm_nn, annot=True, fmt='d', cmap='cividis', cbar=False)
plt.title(f"Confusion Matrix (Accuracy: {accuracy:.2f})")
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```

```python
import xgboost as xgb
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix

dtrain = xgb.DMatrix(X_train, label=Y_train)
dtest = xgb.DMatrix(X_test, label=Y_test)

params = {
    'objective': 'multi:softmax',
    'num_class': 5,
    'max_depth': 5,
    'eta': 0.1,
    'seed': 42
}

num_rounds = 100
bst = xgb.train(params, dtrain, num_rounds)

Y_pred_xgb = bst.predict(dtest)

cm_xgb = confusion_matrix(Y_test, Y_pred_xgb)
print("Confusion Matrix:\n", cm_xgb)
accuracy = accuracy_score(Y_test, Y_pred_xgb)
print("Accuracy Score:", accuracy)

plt.figure(figsize=(8, 6))
sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Reds', cbar=False)
plt.title(f"Confusion Matrix (Accuracy: {accuracy:.2f})")
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from xgboost import XGBClassifier
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your dataset has features 'X' and target 'Y'
# Example: Replace 'feature_columns' and 'target_column' with your actual column names
feature_columns = ['GENDER', 'AGE','SMOKING','YELLOW_FINGERS','ANXIETY','PEER_PRESSURE', 'CHRONIC DISEASE','FATIGUE','ALLERGY','WHEEZING','ALCOHOL CONSUMING', 'COUGHING','SHORTNESS OF BREATH','SWALLOWING DIFFICULTY','CHEST PAIN']  # Replace with your actual feature names
target_column = 'LUNG_CANCER'  # Replace with your actual target column name

# Load your data into a DataFrame (Assuming 'data' is already defined)
X = data[feature_columns]
Y = data[target_column]

# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# Initialize the XGBClassifier with initial parameters
xgb_clf = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    subsample=0.8,
    colsample_bytree=0.8,
    use_label_encoder=False,
    eval_metric='logloss'
)

# Perform Grid Search for hyperparameter tuning
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

grid_search = GridSearchCV(estimator=xgb_clf, param_grid=param_grid, scoring='accuracy', cv=5, verbose=1)
grid_search.fit(X_train, Y_train)

# Best parameters and score
print(f"Best Parameters: {grid_search.best_params_}")
print(f"Best Score: {grid_search.best_score_:.2f}")

# Retrain with best parameters
best_xgb_clf = grid_search.best_estimator_

# Predict on the test set
y_pred = best_xgb_clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(Y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
print("Classification Report:\n", classification_report(Y_test, y_pred))

# Confusion matrix visualization
cm = confusion_matrix(Y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='cividis', cbar=False)
plt.title(f"Confusion Matrix (Accuracy: {accuracy:.2f})")
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```
